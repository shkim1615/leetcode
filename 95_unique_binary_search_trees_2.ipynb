```python
n = 3
nums = [i for i in range(1, n+1)]
print(nums)

subsets = [[]]

for num in nums:
    size = len(subsets)
    for y in range(size):
        subsets.append(subsets[y] + [num])
print(subsets)
```

```python
subsets_separation = [item for item in subsets if len(item) == n]
print(subsets_separation)

# 음.. 부분집합이 아니라 순열이었음..
```

```python
# 이진 트리 만들어 봅시다
# 만들어져 있는데?

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
# 이게 생긴게 이래서. 그냥 여기의 하위 항목으로 함수를 넣는 게 좋아보임아보임
# 이렇게 하면 트리 구조가 생성이 되어 버림. 
# 근데 트리 클래스를 여기서 줌.
# 그러면 트리를 만들어놓고 그 트리를 갖다가 내가 배열로 다시 만들어야할 거 같은데데
def generateTrees(n: int):
    def permutation(arr, n):
        result = []
        if n > len(arr):
            return result
        
        if n == 1:
            for i in arr:
                result.append([i])
        elif n > 1:
            for i in range(len(arr)):
                ans = [i for i in arr]
                ans.remove(arr[i])

                for p in permutation(ans, n-1):
                    result.append([arr[i]] + p)
            
        return result
    
    def generate_root(val):
        root = TreeNode(val=val, left=None, right=None)
        return root
    
    def insert_node(root, val):
        if root == None:
            return generate_root(val)
        
        if val < root.val:
            root.left = insert_node(root.left, val)
        else:
            root.right = insert_node(root.right, val)
        return root
    
    def print_tree(root):
        result = []
        result.append(root.val)
        if root.left is not None:
            tmp = print_tree(root.left)
            for i in tmp:
                result.append(i)
            
            #result.append(print_tree(root.left))
        
        if root.right is not None:
            tmp = print_tree(root.right)
            for i in tmp:
                result.append(i)
            #result.append(print_tree(root.right))
        
        # print(result)
        return result
            
    
    nums = [i for i in range(1, n+1)]
    perm_list = permutation(nums, n)
    print(perm_list)
    
    root = None
    result = []  
    temp = []      
    for arr in perm_list:
        root = None
        for i in arr:
            root = insert_node(root, i)
        result.append(root)
        temp.append(print_tree(root))
    print(result)
    
    for i in range(len(temp)):
        if temp.count(temp[i]) >= 2:
            temp[i] = None
    print(temp)
    
    tmp = [arr for arr in temp if arr is not None]
    print(tmp)
        
        
        
            

            
    
generateTrees(3)
```
